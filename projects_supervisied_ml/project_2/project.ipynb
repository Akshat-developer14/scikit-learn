# Project 2: Predicting House Prices

- Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

- Load the dataset

```python
data = pd.read_csv('student_performance_score.csv')
```

- Input and Output

```python
X = data[['Study_Hours_per_Week']]
y = data['Final_Exam_Score']
```

- Train model

```python
model = LinearRegression()
model.fit(X, y)
```

```python
predicted_score = model.predict(X)

# Evaluate the model

# We will use round() to round the values to 2 decimal places.

print("Evaluation Metrics:")
mae = mean_absolute_error(y, predicted_score)
print(f"Mean Absolute Error: {round(mae, 2)}")

mse = mean_squared_error(y, predicted_score)
print(f"Mean Squared Error: {round(mse, 2)}")

rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {round(rmse, 2)}")

r2 = r2_score(y, predicted_score)
print(f"R^2 Score (Model Accuracy): {round(r2, 4)}")

# Plot the data (Histogram)
plt.figure(figsize=(10, 6))
plt.hist(data['Final_Exam_Score'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Final Exam Scores')
plt.xlabel('Final Exam Score')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()

# Scatter + Regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Scores')
plt.plot(X, predicted_score, color='red', label='Predicted Scores (Regression Line)')
plt.title('Model Prediction vs Actual Scores')
plt.xlabel('Study Hours per Week')
plt.ylabel('Final Output')
plt.grid(True)
plt.show()

new_hours = 9
new_predicted_score = model.predict([[new_hours]])
print(f"Predicted score for {new_hours} hours of study: {new_predicted_score[0]}")
```
